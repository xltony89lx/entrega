```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame simulado
data = {
    'ID': [101, 102, 103, 101, 105],
    'Marca': ['Michelin', 'Pirelli', 'Bridgestone', 'Michelin', 'Firestone'],
    'Medida': ['205/55 R16', '225/45 R17', '195/65 R15', '205/55 R16', '215/60 R16'],
    'Cantidad': [5, 2, 0, 3, 4],
    'Precio Unitario': [1800, 2200, 1500, 1750, 1900],
    'Estado': ['Nueva', 'Nueva', 'Usada', 'Nueva', 'Nueva']
}

df = pd.DataFrame(data)

# Mostrar el inventario
print(" Inventario actual:\n")
print(df)

# Detectar duplicados por ID
print("\n Llantas con ID duplicado:\n")
duplicados = df[df.duplicated('ID', keep=False)]
print(duplicados)

# Detectar productos con stock en 0
print("\n Llantas sin stock:\n")
sin_stock = df[df['Cantidad'] == 0]
print(sin_stock)

# Detectar diferencias de precio para el mismo ID
print("\n Diferencias de precio por mismo ID:\n")
precios_diferentes = df.groupby('ID')['Precio Unitario'].nunique()
precios_conflictivos = precios_diferentes[precios_diferentes > 1]
print(df[df['ID'].isin(precios_conflictivos.index)])

# Sugerencia de limpieza
print("\n Recomendación: Unificar llantas con ID 101 y verificar cuál es el precio correcto.")

# Cantidad de llantas por marca
plt.figure(figsize=(8, 5))
df.groupby('Marca')['Cantidad'].sum().plot(kind='bar', color='skyblue')
plt.title('Cantidad de llantas por marca')
plt.ylabel('Cantidad total')
plt.xlabel('Marca')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Estado de las llantas (nuevas vs usadas)
plt.figure(figsize=(6, 6))
df['Estado'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Distribución del estado de las llantas')
plt.ylabel('')
plt.tight_layout()
plt.show()

# Stock total por medida
plt.figure(figsize=(8, 5))
df.groupby('Medida')['Cantidad'].sum().sort_values().plot(kind='barh', color='goldenrod')
plt.title('Stock total por medida')
plt.xlabel('Cantidad')
plt.ylabel('Medida')
plt.tight_layout()
plt.show()
```
